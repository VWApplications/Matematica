# Método do ponto fixo
***

O problema do ponto fixo consiste em encontrar o valor de $x$ para o qual $f(x) = x$.

* Este problema geralmente ocorre em sistemas dinâmicos em que $f(x)$ calcula o novo estado do sistema a partir do estado anterior $x$.


* Por exemplo, se $x$ representa a posição de um sistema em uma animação $f(x)$ representaria a posição no frame posterior.


* O ponto fixo portanto representaria uma posição estacionária neste sistema: se $f(x) = x$, então o próximo estado é idêntico ao primeiro e assim sucessivamente.


* É lógico que se reorganizarmos um pouco os termos temos $f(x) - x = 0$, que é um problema de encontrar as raízes (zeros) de uma função.

Deste modo, se quisermos encontrar o ponto fixo de uma função (digamos $f(x) = cos(x)$), então podemos utilizar qualquer método de encontrar raízes para a função $g(x) = f(x) - x$.

***
### Exemplos
***

Importar as bibliotecas

```python
import numpy
import matplotlib.pyplot as matplot
%matplotlib inline
```

***

Definimos a nossa função $f(x) = x^2 + 2x + 1$ e o valor de $y_0$

Queremos encontrar o valor de $x$ para o qual $f(x) = y_0$

```python
def f(x):
    return x*x + 2*x + 1

y0 = 42
```

***

Definimos $g(x)$, a função que queremos encontrar os zeros

```python
def g(x):
    return f(x) - y0
```

***

Também podemos encontrar um ponto fixo explicitamente aplicando $f(x)$ sucessivamente nos valores de x

```python
values = [1]
x = 1
for i in range(20):
    x = numpy.cos(x)
    values.append(x)
    print('%2d) x: %.7f' % (i, x))
    
matplot.plot(values)
matplot.xlabel('número de iterações')
matplot.ylabel('x')
matplot.show()
```

***

Podemos visualizar os dados anteriores em um gráfico com a função

```python
points = [(1, 0)]
x = 1
for i in range(20):
    x, x_old = numpy.cos(x), x
    # Ponto na funçao cos(x)
    points.append((x_old, x))

    # Ponto projetado na reta y = x
    points.append((x, x))

print(points)
# Desenha y = cos(x) e y = x
X = numpy.linspace(0, 2)
Y = numpy.cos(X)
matplot.plot(X, Y, 'b-', label='cos(x)')
matplot.plot(X, X, 'k-')

# Desenha mapa de iteração
X, Y = numpy.array(points).T
matplot.plot(X, Y, 'r-', label='x_n')

# Define os limites do gráfico [xmin, xmax, ymin, ymax]
matplot.axis([0.5, 1.1, 0.4, 1.0]) 

# Elementos do gráfico
matplot.legend(loc='best')
matplot.ylabel('cos(x)')
matplot.xlabel('x')
matplot.show()
```

***
### Método do ponto fixo
***

Vimos que é possível resolver um problema de ponto fixo como um problema de encontrar raízes. Também é possível fazer o contrário e resolver um problema de raízes como se fosse um ponto fixo. Para isto, considere a que queremos encontrar as raízes tais quais $g(x) = 0$. Com um pouquinho de álgebra, temos $x + kg(x) = x$, o que pode ser reescrito como $h(x) = x$, onde definimos $h(x) = x + kg(x)$ e $k$ é uma constante arbitrária.

Deste modo, encontrar a raiz de $g(x)$ é equivalente a encontrar o ponto fixo de $h(x) = x + kg(x)$.

```python
def h(x):
    return x - 0.1 * g(x)  # definimos k = -0.1
```

***

```python
x = 5
points = []
for i in range(20):
    x, x_old = h(x), x
    
    # Ponto na funçao h(x)
    points.append((x_old, x))

    # Ponto projetado na reta y = x
    points.append((x, x))

print(points)
# Desenha y = g(x) + x e y = x
X = numpy.linspace(0, 10)
Y = h(X)
matplot.plot(X, Y, 'b-', label='x - g(x)')
matplot.plot(X, X, 'k-')

# Desenha mapa de iteração
X, Y = numpy.array(points).T
matplot.plot(X, Y, 'r-', label='x_n')

# Define os limites do gráfico [xmin, xmax, ymin, ymax]
matplot.axis([5, 6, 5, 6]) 

# Elementos do gráfico
matplot.legend(loc='best')
matplot.ylabel('h(x)')
matplot.xlabel('x')
matplot.show()
```

***
### Função do ponto fixo
***

```python
def fixed_point(g, x0, k=1, x_tolerance=1e-6, y_tolerance=1e-6):
    """
    Calcula o zero da função g(x) usando o método do ponto fixo e iterando 
    sobre a função h(x) = x + k * g(x)
    
    Argumentos:
    
        g: uma função de uma única variável
        x0: valor inicial
        k: valor usado para criar h(x) = x + k * g(x)
        x_tolerance: tolerância em x (retorna quando intervalo for menor que x_tolerance)
        y_tolerance: tolerância em y (retorna quando |g(x)| < y_tolerance)
    
    Retornos:
    
        Retorna um zero de g(x) (valor de x em que g(x) = 0).
    """
    x = x0
    
    while True:
        value = g(x)
        x_new = x + k * value
        x, dx = x_new, x_new - x
        if abs(dx) < x_tolerance or abs(value) < y_tolerance:
            break
        x = x_new
    return x
```

***

```python
print("x =", fixed_point(g, 5, k=-0.08))
print("x =", fixed_point(numpy.cos, 1))
```

***

O ponto fixo é considerado um método bastante rápido. No entanto, é um dos métodos mais frágeis dos considerados. Em muitos casos não há convergência ou a velocidade de convergência pode ser fortemente dependente do parâmetro k. Como regra geral nunca use o método do ponto fixo!. A única exceção é quando é possível provar que a função que você deseja encontrar a raiz possui as propriedades necessárias para que o método funcione.
